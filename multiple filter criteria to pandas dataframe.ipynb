```python
import pandas as pd
```

```python
df=pd.read_csv('https://bit.ly/imdbratings')
```

```python
df.head()
```

```python
df[df.duration>=200]
```

```python
## or operator with multiple filter 
df[(df.duration>=200) | (df.genre=='Drama')]
```

```python
## And operator with mutiple filter
df[(df.duration>=200) & (df.genre=='Drama')]
```

```python
# multiple columns
df[(df.genre=='Crime')|(df.genre=='Drama')|(df.genre=='Adventure')|(df['star_rating']==8.7)]
```

```python
## Another sortcut method
df[df['duration'].isin([154])]
```
